```
from tensorflow.keras.utils import to_categorical
import numpy as np

train_x, train_y = df_2.iloc[:,:22 ], df_2.iloc[:,22]
test_x, test_y = df_2.iloc[:,:22], df_2.iloc[:,22]
# 1. normalization
train_X = (train_x - train_x.min()) / (train_x.max() - train_x.min())
train_X = train_X.to_numpy()
test_X = (test_x - test_x.min()) / (test_x.max() - test_x.min())
test_X = test_X.to_numpy()
# 2. label vectorization
train_Y = to_categorical(train_y, 12)
test_Y = to_categorical(test_y, 12)
```

```
def generate(x, y, timestep):
     x_train = []
     y_train = []
     for i in range(len(x)):
         temp_x = x[i:(i + timestep)]
         if (i + timestep) < len(x):
             temp_y = y[i + timestep]
             x_train.append(temp_x)
             y_train.append(temp_y)
         else:
             break
     return np.array(x_train), np.array(y_train)
```

```
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt


model = Sequential()
model.add(LSTM(32, input_shape=(20,22), return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(16))
model.add(Dropout(0.2))
model.add(Dense(12, activation='softmax'))

early_stopping = EarlyStopping(monitor='val_loss', patience=10, mode='min')

check_point = ModelCheckpoint(filepath='/content/drive/My Drive/project/코드/best_model.h5',
                                monitor='val_loss',
                                save_best_only=True)

model.compile(loss='categorical_crossentropy', optimizer='adam',
              metrics=['accuracy'])

history = model.fit(train_X_2, train_Y_2, batch_size=64, epochs=200,
                    validation_data=(test_X_2, test_Y_2),
                    class_weight=class_weights,
                    callbacks=[early_stopping, check_point])

```

```
test_y_loss = history.history['val_loss']
train__yloss = history.history['loss']

test_y_acc = history.history['val_acc']
train__y_acc = history.history['acc']

x_len = np.arange(len(test_y_loss))

fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()


loss_ax.plot (x_len, test_y_loss, marker=',', c='red', label='Testset_loss')
loss_ax.plot(x_len, train__yloss, marker=',', c='blue', label='Trainset_loss')
loss_ax.set_ylabel('loss')
loss_ax.legend(loc="upper left")

acc_ax.plot (x_len, test_y_acc, marker=',', c='green', label='Testset_acc')
acc_ax.plot(x_len, train__y_acc, marker=',', c='black', label='Trainset_acc')
acc_ax.set_ylabel('loss')


plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.show()
```
