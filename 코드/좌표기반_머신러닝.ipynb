```
from tensorflow.keras.utils import to_categorical
import numpy as np

train_x, train_y = df_all.iloc[:40000,:22], df_all.iloc[:40000,22]
test_x, test_y = df_all.iloc[40000:,:22], df_all.iloc[40000:,22]
# 1. normalization
train_X = (train_x - train_x.min()) / (train_x.max() - train_x.min())
train_X = train_X.to_numpy()
test_X = (test_x - test_x.min()) / (test_x.max() - test_x.min())
test_X = test_X.to_numpy()
# 2. label vectorization
train_Y = to_categorical(train_y, 18)
test_Y = to_categorical(test_y, 18)
```

```
def generate(x, y, timestep):
     x_train = []
     y_train = []
     for i in range(len(x)):
         temp_x = x[i:(i + timestep)]
         if (i + timestep) < len(x):
             temp_y = y[i + timestep]
             x_train.append(temp_x)
             y_train.append(temp_y)
         else:
             break
     return np.array(x_train), np.array(y_train)
```

```
g = generate(train_X, train_Y, 80)
g_2 = generate(test_X, test_Y, 80)
```

```
print(g[0].shape)
print(g[1].shape)
print(g_2[0].shape)
print(g_2[1].shape)
```

```
train_X_2, train_Y_2 = g[0], g[1]
test_X_2, test_Y_2 = g_2[0], g_2[1]
```

```
from sklearn.utils import class_weight

class_weights = class_weight.compute_class_weight('balanced',
                                                 np.unique(train_y),
                                                 train_y)
```

```
class_weights
```

```
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, GRU, MaxPooling1D, Conv1D
from keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt


model3 = Sequential()
model3.add(Conv1D(256, 5, activation='relu',
                        input_shape=(80, 22)))
model3.add(MaxPooling1D(3))
model3.add(GRU(64, recurrent_dropout=0.2, return_sequences=True,
               kernel_initializer="he_uniform"))
model3.add(GRU(32, recurrent_dropout=0.2,
               kernel_initializer="he_uniform"))
model3.add(Dense(18, activation='softmax'))

model3.compile(loss="categorical_crossentropy", optimizer='adam', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, mode='min')

check_point = ModelCheckpoint(filepath='/content/drive/My Drive/project/코드/best_model.h5',
                                monitor='val_loss',
                                save_best_only=True)

model3.summary()

```

```
history = model3.fit(train_X_2, train_Y_2, batch_size=256, epochs=50,
                    validation_data=(test_X_2, test_Y_2),
                    class_weight=class_weights,
                     callbacks=[check_point, early_stopping])

test_y_loss = history.history['val_loss']
train__yloss = history.history['loss']

x_len = np.arange(len(test_y_loss))

plt.plot (x_len, test_y_loss, marker=',', c='red', label='Testset_loss')
plt.plot(x_len, train__yloss, marker=',', c='blue', label='Trainset_loss')

plt.legend(loc='upper right')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```
