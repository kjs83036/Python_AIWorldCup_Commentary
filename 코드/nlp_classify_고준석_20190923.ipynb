```
class fbData():

    def __init__(self, df, count, result):
        self.df = df.text
        self.count = count
        self.result = result
        print(self.df.iloc[0])
    
    def show_sentense(self, word):
        self.df = self.df.apply(lambda x: None if word in x else x)
        self.df = self.df.dropna()
        
        if self.df.empty == True:
            print("empty")
        else:
            self.df = self.df.apply(lambda x: None if word in x else x)
            self.df = self.df.dropna()
            print(self.df.iloc[0])

    def save_result(self, words, robot):

        self.result.append([self.df.iloc[0], words, robot])
        self.count += 1
```

```
class fbData_2(fbData):
    def __init__(self, df, count, result, n):
        super().__init__(df, count, result)
        for i in range(1, n):
            print(self.df.iloc[i])
    def show_sentense(self, word, n):
        super().show_sentense(word)
        if self.df.empty == True:
            print("empty")
        else:
            for i in range(1, n):
                print(self.df.iloc[i])
    def save_result(self, words, robot, n):
        self.result.append([])
        for i in range(0, n):
            self.result[self.count].append([self.df.iloc[i], words[i], robot[i]])
        self.count += 1
```
